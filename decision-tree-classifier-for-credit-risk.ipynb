```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../input/default-of-credit-card-clients-dataset/UCI_Credit_Card.csv')
data.shape
```

```python
# Renomeando coluna
data = data.rename(columns={'PAY_0': 'PAY_1'})
```

```python
# Removendo valores -2
data = data[data.PAY_1 != -2]
data = data[data.PAY_2 != -2]
data = data[data.PAY_3 != -2]
data = data[data.PAY_4 != -2]
data = data[data.PAY_5 != -2]
data = data[data.PAY_6 != -2]

# Substituindo -1 por 0
np.where(data['PAY_1'] == -1, 0, data['PAY_1'])
np.where(data['PAY_2'] == -1, 0, data['PAY_2'])
np.where(data['PAY_3'] == -1, 0, data['PAY_3'])
np.where(data['PAY_4'] == -1, 0, data['PAY_4'])
np.where(data['PAY_5'] == -1, 0, data['PAY_5'])
np.where(data['PAY_6'] == -1, 0, data['PAY_6'])

data.shape
```

```python
# Atribuindo coluna alvo
def apply_sum(row):
    return row[{'PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6'}].sum()

target = [apply_sum(row) for idx, row in data.iterrows()]
data['TARGET'] = target
```

```python
# Definindo score de risco de crédito
#data['TARGET'] = np.where(data['TARGET'] < 1, 1, data['TARGET'])
#data['TARGET'] = np.where(data['TARGET'].between(1,2), 2, data['TARGET'])
#data['TARGET'] = np.where(data['TARGET'].between(3,4), 3, data['TARGET'])
#data['TARGET'] = np.where(data['TARGET'].between(5,6), 4, data['TARGET'])
#data['TARGET'] = np.where(data['TARGET'] > 6, 5, data['TARGET'])
```

```python
# Removendo linhas desnecessárias
data = data[data.EDUCATION != 0]
data = data[data.EDUCATION != 5]
data = data[data.EDUCATION != 6]
data = data[data.MARRIAGE != 0]
data.shape
```

```python
data.to_csv(r'FileName.csv')
```

```python
# Aplicando modelo
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.model_selection import train_test_split

y = data['TARGET'].copy()
features = ['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_1', 'PAY_2',
       'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2',
       'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1',
       'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
X = data[features].copy()
X.columns

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

classifier = DecisionTreeClassifier(max_depth=10, random_state=14) 
classifier.fit(X_train, y_train)
predictions = classifier.predict(X_test)
accuracy_score(y_true = y_test, y_pred = predictions)
```
